```python
import pandas as pd
```

```python
df = pd.read_csv('data/datos_ejemplo.csv')
```

```python
df.head()
```

```python
# ------------------------------
# Manage null values
# ------------------------------

print("nulls values")
df.isnull().sum()

# delete rows with null values
df_without_nulls = df.dropna()

# refill null values
df_refill = df.fillna({'salario': 0, 'Nombre': 'Desconocido'})

print("\ndata without nulls")
print(df_without_nulls)

print("\nrefill null values")
print(df_refill)

```

```python
# ------------------------------
# detect and fix errors in data types
# ------------------------------

df['Edad'] = pd.to_numeric(df['Edad'], errors='coerce')

print("\nData types")  
print(df.dtypes)

df['Edad'] = df['Edad'].fillna(df['Edad'].mean())

print("\nData types after the correltion")  
print(df.dtypes)

print("fixed dataframe")
print(df)
```

```python
# -------------------------------------------
# Convert from category to number variable
# -------------------------------------------

df['Genero'] = df['Genero'].map({'F': 0, 'M': 1})

# use get_dummis to convert 'Departamento' to dummy variable
df = pd.get_dummies(df, columns=['Departamento'], drop_first=True)

print("\nDataframe with converted variables: ")
df.head()
```
